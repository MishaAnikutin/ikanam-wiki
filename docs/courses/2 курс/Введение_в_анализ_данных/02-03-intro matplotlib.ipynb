```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 6)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1) 
ax2 = fig.add_subplot(2, 2, 2) 
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), linestyle='--', color='g');
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1) 
ax2 = fig.add_subplot(2, 2, 2) 
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), linestyle='--')
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30));
```

```python
plt.plot(np.random.randn(30).cumsum(), color='k', linestyle='dashed', marker='o');
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, linestyle='--', label='Default')
plt.plot(data, linestyle='-', drawstyle='steps-post', label='steps-post');
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum(), label='синий график')
ax.plot(np.random.randn(1000).cumsum(), label='оранжевый график')

ax.set_xticks([0, 250, 500, 750, 1000])
ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=45, fontsize='small')

ax.legend(loc='best')
ax.set_title('Мой график')
ax.set_xlabel('Шаги');
```
